```python
import matplotlib.pyplot as plt

# 
methods = ['Crossentropy Loss', 'Crossentropy + MMD Loss', 'Crossentropy + XE_Image', 'Crossentropy + MMD Loss + XE_Image']
novel_scores = [73.90, 76.13, 78.60, 79.39]
base_scores = [91.94, 93.36, 95.41, 96.3]

# 
bar_width = 0.35

# 
fig, ax = plt.subplots()
index = range(len(methods))

bar1 = ax.bar(index, base_scores, bar_width, label='Base')
bar2 = ax.bar([i + bar_width for i in index], novel_scores, bar_width, label='Novel')

# 
ax.set_xlabel('Methods')
ax.set_ylabel('Scores')
ax.set_title('Performance Comparison')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(methods, rotation=45, ha='right')
ax.legend()

# 
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bar1)
autolabel(bar2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 
methods = ['Crossentropy Loss', 'Crossentropy + MMD Loss', 'Crossentropy + XE_Image', 'Crossentropy + MMD Loss + XE_Image']
novel_scores = [73.90, 76.13, 78.60, 79.39]
base_scores = [91.94, 93.36, 95.41, 96.3]

# 
bar_width = 0.35

# 
fig, ax = plt.subplots()
index = range(len(methods))

bar1 = ax.bar(index, novel_scores, bar_width, label='Novel')
bar2 = ax.bar([i + bar_width for i in index], base_scores, bar_width, label='Base')

# 
ax.set_xlabel('Methods')
ax.set_ylabel('Scores')
ax.set_title('Performance Comparison')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(methods, rotation=45, ha='right')
ax.legend(title='Groups')

# 
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bar1)
autolabel(bar2)

plt.tight_layout()
plt.show()
```
