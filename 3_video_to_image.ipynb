```python
import cv2 as cv
import os

'''Get video information'''
file = "input/test.mp4"
save_file = "output/" + str(file).split("/")[1].split(".mp4")[0] + "/origin/"
print(save_file)

cap=cv.VideoCapture(file) #load video
fps=cap.get(cv.CAP_PROP_FPS) # get frame rate
width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH)) # get width
height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT)) # get height
print(fps,width,height)

'''video to pictures'''
isOpened=cap.isOpened()
i=0
while(isOpened):
    i=i+1
    flag,frame=cap.read()
    # if 47 < i and i < 48:
    fileName = '%03d'%i+".jpg"
    print(fileName)
    if flag == True :
        if os.path.exists(save_file) is False:
            os.makedirs(save_file)
        # y1 = bbox[0] 13
        # x1 = bbox[1] 706
        # y2 = bbox[2] 96
        # x2 = bbox[3] 797
        # cv.rectangle(frame,(706, 13), (797, 96), (88, 87, 86), -1) 
        cv.imwrite(save_file +'%03d'%i+".jpg",frame) # name the image
        cv.waitKey(1)
    else:
        break
cap.release()
print('end')

```
